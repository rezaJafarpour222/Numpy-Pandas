# Array Basics
- array.shape --> shape of array.

- array.size --> size of array.

- array.ndim --> array dimension.

- array.T.shape -->  shape of array after transpose.

- array.reshape(x, y) -->  reshape array to a new array with(x, y) dimensions.


```python
print(array.shape)
print(array2d.size)
print(array2d.T.shape)
print(array2d.ndim)
print(array2d.reshape(2,5))
```

# Array Creation
- np.ones((rows, cols), dtype) --> Creates an array of ones of a given size(by default dtype == float).

- np.zeros((rows, cols), dtype) --> Creates an array of ones of a given size(by default dtype == float).

- np.arange((start, stop), step) --> Creates an array of integers with given start & stop values, and a step size(only stop is required, and is not inclusive).

- np.linspace((start, stop), n) --> Creates an array of floats with given start & stop values with n elements, seperated by a consistent tep size   (stop is inclusive).


```python
ones = np.ones(4,)
print(f" zeros---> {ones}\n")
zeros = np.zeros((2, 5), dtype=int)
print(f" ones---> {zeros}\n")
arange = np.arange(10)
print(f" arange---> {arange}\n")
linspace = np.linspace(0, 100, 5)
print(f" linspace---> {linspace}\n"  )
example = np.arange(1, 9 ,2).reshape(2,2)
print("-->generating odd numbers from 1 to 9 (step =2), reshape into 2x2 array\n")
print(f" example---> {example}\n")
identity = np.identity(10)
print(f"identity matrix ---> {identity}")
# s = np.linspace(0, 100 , 10) since '0' is included , step must be '11' and not '10' to get 0, 10, 20 ,...
s = np.linspace(0, 100 , 11)
print(s)
```

# Random Number Arrays
## in order to generate random number we need random generator
- np.default_rng(seed) --> Creates random number generator(the seed is for reproducibility).

- rng.random(n) --> Returns n random numbers from a uniform distribution between 0 and 1.

- rng.normal(mean, stdev, n) --> Returns n random number from a normal distribution with a given mean and standart deviation.



```python
from numpy.random import  default_rng
```

```python
rng = default_rng(12345) #seed = 12345
random_arr = rng.random(10)
random_arr
```

```python
mean, stdev = 5, 1
random_normal = rng.normal(mean, stdev, size=10)
random_normal
```

```python
integers = rng.integers(0, 10, 100)
integers
```

# Array operations


```python
rand = np.random.default_rng(1234)
inventory = rand.integers(0, 100, 10)
price = (rand.random(10) * 10).round(2)
(inventory * price).sum()
```
